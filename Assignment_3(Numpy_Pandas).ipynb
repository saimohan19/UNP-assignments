2.In the “dat.csv” replace all the null values with the mean of their respective columns (except Gender).


```python
data=pd.read_csv("dat.csv",index_col=0)
data.isna()# gives boolean value to check the cell is empty or not
data["class"].fillna(np.mean(data["class"]),inplace=True)
data["Age"].fillna(np.mean(data["Age"]),inplace=True)
data["Survived"].fillna(np.mean(data["Survived"]),inplace=True)
data
```
